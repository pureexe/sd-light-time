```python
import os
import torch
import matplotlib.pyplot as plt
```

```python
light0_inv = torch.load('../../output/20241012_backforth/val_backforth_v2/1.0/vae/1.0/depth/1e-4/chk199/lightning_logs/version_0/ddim_latents/everett_dining1-dir_0_mip2.pt',map_location=torch.device('cpu'))
light21_inv = torch.load('../../output/20241012_backforth/val_backforth_v2/1.0/vae/1.0/depth/1e-4/chk199/lightning_logs/version_0/ddim_latents/everett_dining1-dir_21_mip2.pt',map_location=torch.device('cpu'))
```

```python
light0_21_gen = torch.load('../../output/20241012_backforth/val_backforth_v2/1.0/vae/1.0/depth/1e-4/chk199/lightning_logs/version_0/ddim_latents/everett_dining1-dir_0_mip2_everett_dining1-dir_21_mip2.pt',map_location=torch.device('cpu'))
light21_0_gen = torch.load('../../output/20241012_backforth/val_backforth_v2/1.0/vae/1.0/depth/1e-4/chk199/lightning_logs/version_0/ddim_latents/everett_dining1-dir_21_mip2_everett_dining1-dir_0_mip2.pt',map_location=torch.device('cpu'))
```

```python
# find mean over list
light0_inv_mean = torch.stack(light0_inv).view(len(light0_inv),-1).mean(dim=1)
light21_inv_mean = torch.stack(light21_inv).view(len(light21_inv),-1).mean(dim=1)
light0_21_gen_mean = torch.stack(light0_21_gen).view(len(light0_21_gen),-1).mean(dim=1)
light21_0_gen_mean = torch.stack(light21_0_gen).view(len(light21_0_gen),-1).mean(dim=1)
```

```python
# load timesteps
timesteps0 = torch.load('../../output/20241012_backforth/val_backforth_v2/1.0/vae/1.0/depth/1e-4/chk199/lightning_logs/version_0/ddim_latents/everett_dining1-dir_0_mip2_timesteps.pt',map_location=torch.device('cpu'))
timesteps21 = torch.load('../../output/20241012_backforth/val_backforth_v2/1.0/vae/1.0/depth/1e-4/chk199/lightning_logs/version_0/ddim_latents/everett_dining1-dir_21_mip2_timesteps.pt',map_location=torch.device('cpu'))
```

```python
# plot the mean of the inverse depth
plt.plot(timesteps0,light0_inv_mean,label='light_0_inv',color='red')
plt.plot(timesteps21,light21_inv_mean,label='light_21_inv', color='green')
plt.plot(timesteps0[::-1],light0_21_gen_mean,label='gen_0_to_21',linestyle='dashed', color='red')
plt.plot(timesteps21[::-1],light21_0_gen_mean,label='gen_21_to_0',linestyle='dashed', color='green')
plt.legend()
plt.xlabel('timesteps')
plt.ylabel('mean of latent')
plt.title('(everett_dining1) mean of latent over timesteps')
plt.show()
```
