```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

```

```python
prompt = "a photo of a perfect mirrored reflective chrome ball sphere placing on beach"
image = pipe(prompt, guidance_scale=3.0, generator=torch.Generator().manual_seed(42)).images[0]  
display(image)
```
