```python
import os 
import torch
import torchvision
```

```python
ROOT_DIR = "../../output/20240902/val_shoe/1.0"
CHECKPOINTS = range(0, 90, 10)
LRS = ['1e-4', '1e-5']
METHODS = ['no_control', 'depth', 'normal', 'both']
directory_template = "{}/{}/chk{}/lightning_logs/version_0/crop_image"
output_dir_template = "frame/{}"
```

```python
for frame_id in range(0, 60):
    for lr in LRS:
        images = []
        for method in METHODS:
            for chk in CHECKPOINTS:
                directory = directory_template.format(method, lr, chk)
                try:
                    file_path = os.path.join(ROOT_DIR, directory, f'00000_{frame_id:05d}.jpg')
                    image = torchvision.io.read_image(file_path).float() / 255
                    # resize to 256x256
                    image = torchvision.transforms.functional.resize(image, (256, 256))
                except:
                    image = torch.zeros(3, 256, 256)
                images.append(image)
        output_dir = output_dir_template.format(lr)
        os.makedirs(os.path.join(ROOT_DIR, output_dir), exist_ok=True)
        output_path = os.path.join(ROOT_DIR, output_dir, f'{frame_id:05d}.jpg')
        print(output_path)
        # make grid using torchvision
        grid = torchvision.utils.make_grid(images, nrow=len(CHECKPOINTS))
        torchvision.utils.save_image(grid, output_path)
```
