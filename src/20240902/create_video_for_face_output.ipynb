```python
import os 
```

```python
#CONTROLS = ['both','depth']
CONTROLS = ['depth']
GUIDANCES = ['1.0','3.0','5.0','7.0']
```

```python

for control in CONTROLS:
   for guidance in GUIDANCES:
        root_dir = f'../../output/20240902/val_face_roll60_v2_scene2/{guidance}/{control}/1e-4/chk319/lightning_logs/version_0'
        input_dir = root_dir + '/crop_image'
        output_dir = root_dir + f'/video_crop'
        os.makedirs(output_dir, exist_ok=True)
        # create directory
        for face_id in range(10):
            if os.path.exists(f'{output_dir}/{face_id:04d}.mp4'):
                continue
            cmd = f'ffmpeg -r 6 -i "{input_dir}/{face_id:05d}_02%03d.jpg" -c:v libx264 -crf 12 -pix_fmt yuv420p {output_dir}/{face_id:04d}.mp4'
            #print(cmd)
            os.system(cmd)
```
