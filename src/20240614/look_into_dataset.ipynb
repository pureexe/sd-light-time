```python
from FaceSingleAxisDataset import FaceSingleAxisDataset
```

```python
import torch 
from matplotlib import pyplot as plt
import numpy as np
```

```python
train_dataset = FaceSingleAxisDataset(split="train", root_dir="../../datasets/face/face2000_single")
```

```python
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=1, shuffle=False)
```

```python
lights = []
for i, data in enumerate(train_loader):
    lights.append(data['light'].numpy())
```

```python
lights = np.array(lights).squeeze()
```

```python
lights.min()
```

```python
lights.max()
```

```python
plt.hist(lights, bins=20, range=(-1, 1))
plt.title('Distribution of Values in [-1, 1]')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
